```python
import csv

dates = np.array(["2014-12-27", "2015-03-28", "2015-06-25", "2015-09-26",
                 "2015-12-26", "2016-03-26", "2016-06-25", "2016-09-24",
                 "2016-12-31", "2017-04-01", "2017-07-01", "2017-09-30",
                 "2017-12-30", "2018-03-31", "2018-06-30", "2018-09-29",
                 "2018-12-29", "2019-03-30", "2019-06-29", "2019-09-28",
                 "2019-12-28", "2020-03-28", "2020-06-27", "2020-09-26"])

# dict = {"Date": ) }
headers = np.array(["Change in Net Income", "Change_Cash_Operations", "Change_Cash_Investments", "Change_Cash_Financials", "Change_Cash_Equivalents"])

####### Annotation: This code is from the docs.python.org csv page.
####### Annotation: https://docs.python.org/3/library/csv.html 

quarterly_datasets = {}
with open('Quarterly_Data.csv', newline='') as q_data:
  q_reader = csv.DictReader(q_data)
  i = 0
  for row in q_reader:
    q_dataset = np.array([row['Change in Net Income'], row['Change_Cash_Operations'], row['Change_Cash_Investments'], row['Change_Cash_Financials'], row['Change_Cash_Equivalents']])
    quarterly_datasets[dates[i]] = q_dataset
    i = i + 1

df = pd.DataFrame(quarterly_datasets).T
df.columns = headers

df
```

```python
#Retrieve adjusted close for each date we have
dates = list(quarterly_datasets.keys())
close = np.array([])
start = ""
apple = yf.Ticker("AAPL")

# yf.download("AAPL", start = dates[22], end = dates[22])["Adj Close"]
for i in range(len(dates) - 1):
    start_date = dates[i]
    end_date = dates[i + 1]
    adj = yf.download("AAPL", start = start_date, end = end_date)["Adj Close"]
    close = np.append(close, adj[0])

    if i is len(dates) - 2:
        close = np.append(close, adj[-1])
```
